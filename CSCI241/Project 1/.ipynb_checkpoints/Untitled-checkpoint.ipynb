```python
import time 
import random

def insertion_sort(arr):
    for k in range(1, len(arr)):
        cur = arr[k]
        j = k
        while j > 0 and arr[j-1] > cur:
            arr[j] = arr[j-1]
            j = j - 1
        arr[j] = cur

def selection_sort(arr):
    for k in range(len(arr)):
        min_arr = k 
        j = k 
        while j+1 in range(len(arr)):
            j = j+1
            if arr[min_arr] > arr[j]:
                min_arr = j 
    
        temp = arr[k]
        arr[k] = arr[min_arr]
        arr[min_arr] = temp
        
            
```
